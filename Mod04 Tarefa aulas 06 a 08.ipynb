E podemos ver que temos um limite de 5000 instâncias preenchidas nessa API, no resultado de "links" vemos o a chave "first" e "next" que indicam paginação.

Portanto, na primeira página não conseguimos pegar todos os dados para uf == sp e tipo_interessado == Estado e é preciso fazer esse processo num loop for usando o parâmetro 'next'

````python
print(url.json()["hasMore"])
#print(url.json()["links"])

for link in url.json()["links"]:
    print(link)
    if link['rel'] == 'next':
            next_link = link['href']

print(next_link)
````

````python
# Aprimorando a função

def get_state(page1, uf, tipo_interessado = None):
    url = requests.get(page1)
    df = pd.DataFrame()
    
    temp_df = pd.DataFrame(url.json()['items'])
    df = pd.concat([df, temp_df], ignore_index=True)
        
    while url.json()["hasMore"]:
        
        for link in url.json()["links"]:
            if link['rel'] == 'next':
                next_link = link['href']
                print(next_link)
        url = requests.get(next_link)
        
        temp_df = pd.DataFrame(url.json()['items'])
        df = pd.concat([df, temp_df], ignore_index=True)
        
    if tipo_interessado == None:
        cond = (df["uf"] == uf)
    else:
        cond = (df["uf"] == uf) & (df["tipo_interessado"] == tipo_interessado)
    return df[cond]

url = "https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl"

df_filter = get_state(url, "SP", "Estado")
display(df_filter.info())
display(df2.info())
````

Agora, nossa função get_state retorna os valores corretamente
